# This notebook implements Fourier Series
Sal Khan's lecture on Fourier series computes the Fourier series for the square wave.  This notebook creates similar plots for a spike.  It is really for a dft.


```python
# import some standard libraries
import sys
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

print ("sys.version=",sys.version)
```

```python
# create and plot a square wave

time=np.arange(0.0,2.0*np.pi,2.0*np.pi/32.0)
squarewave=np.zeros(time.shape)
squarewave[0]=1.0
print ('time=',time)
print ('squarewave=',squarewave)
plt.plot(time,squarewave)
plt.show()
```

```python
# nterms is number of sines and cosines
nterms=24
a=np.zeros(nterms)
b=np.zeros(nterms)

a[:]=1.0/nterms
terms=np.arange(nterms)
print('a=',a)
print('b=',b)
```

```python
series_tot=np.zeros(time.shape)

for iterm in range(0,nterms,1):
    series_term=a[iterm]*np.cos(iterm*time)
    series_tot+=series_term
    fig, ax = plt.subplots(figsize=(10, 1))
    ax.plot(time,series_term)
    plt.xlabel("time")
    plt.title("cos(%d*time)"%iterm)
    plt.show()
fig, ax = plt.subplots(figsize=(10, 1))
ax.plot(time,series_tot)
plt.xlabel("time")
plt.title("series_tot")
plt.show()

```
