## Python notebook to read and plot Washington, DC Station 5 seismometer spectra

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

### Read spectra from .txt file with pandas read_csv
- comma separated values, csv
- no headers
- delimited with blanks instead of comma
- ssr_dc05 is a pandas dataframe.  head function prints the first lines

```python
ssr_dc05 = pd.read_csv("SSR.DC05.txt",header=None,names=['freq','SSR'],delim_whitespace=True)
ssr_dc05.head()
```

### Read and print beginnning of the other spectra

```python
hvsr_dc05 = pd.read_csv("HVSR.DC05.txt",header=None,names=['freq','HVSR'],delim_whitespace=True)
hvsr_dc05.head()
```

### Make an inital plot

```python
plt.plot(hvsr_dc05['HVSR'] )
plt.show()
```

### Plot small frequency range with xscale('log')

```python
ax=plt.subplot(111)
ax.plot(hvsr_dc05['freq'],hvsr_dc05['HVSR'] )
ax.set_xlim([1,11.])
ax.set_xscale("log")
plt.show()
```

### Overplot the two spectra

```python
ax=plt.subplot(111)
ax.plot(hvsr_dc05['freq'],hvsr_dc05['HVSR'],ssr_dc05['freq'],ssr_dc05['SSR'] )
ax.set_xlim([0.1,11.])
ax.set_xscale("log")
plt.show()
```

### Plot smaller frequency range, match Excel color, use dots.

```python
ax=plt.subplot(111)
ax.plot(hvsr_dc05['freq'],hvsr_dc05['HVSR'],'ro',ssr_dc05['freq'],ssr_dc05['SSR'],'bo' )
ax.set_xlim([1,3.5])
ax.set_xscale("log")
plt.show()
```

### Crossplot HVSR and SSR

```python
ax=plt.subplot(111)
fmax=3.5
x=hvsr_dc05.loc[hvsr_dc05['freq']<fmax]['HVSR']
y=ssr_dc05.loc[ssr_dc05['freq']<fmax]['SSR']
ax.plot(x,y,'bo' )
plt.show()
```
